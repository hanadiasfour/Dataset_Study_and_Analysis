# Welcome to Video Game Sales Dataset Analysis

## <font color='blue'>Description of the Video Game Sales Dataset:</font>

This dataset contains information about video games with sales greater than 100,000 copies. There are 11 features and 16,598 records. It was generated by scraping data from vgchartz.com. The source of this dataset is Kaggle.com under the title "Video Game Sales". Link to dataset: https://www.kaggle.com/datasets/gregorut/videogamesales 


### <font color='blue'> Dictionary of Columns:</font>

1. **Rank**
   - Data Type: Integer
   - Description: Ranking of overall sales.

2. **Name**
   - Data Type: String
   - Description: The name of the game.

3. **Platform**
   - Data Type: String
   - Description: The platform on which the game was released.

4. **Year**
   - Data Type: Integer
   - Description: The year of the game's release.

5. **Genre**
   - Data Type: String
   - Description: The genre of the game.

6. **Publisher**
   - Data Type: String
   - Description: The publisher of the game.

7. **NA_Sales**
   - Data Type: Float
   - Description: Sales in North America (in millions).

8. **EU_Sales**
   - Data Type: Float
   - Description: Sales in Europe (in millions).

9. **JP_Sales**
   - Data Type: Float
   - Description: Sales in Japan (in millions).

10. **Other_Sales**
    - Data Type: Float
    - Description: Sales in the rest of the world (in millions).

11. **Global_Sales**
    - Data Type: Float
    - Description: Total worldwide sales (in millions).

### <font color='blue'>Based on intuition, the two features NA_Sales (North America Sales) and EU_Sales (Europe Sales) could be correlated.</font>

Logically, if a certain game is popular(has high sales) among the North American audiance, then the Europian people will also like this game and therefore its sales there will increase too. In other words, I expect the NA_Sales and EU_Sales features to have a positive relation.

### <font color='blue'>To verify whether this is true, two methods are used:</font>
- <font color='blue'>Visualize relation</font>

```python
# We can visualize the relation by using a scatter plot
data.plot.scatter(x='EU_Sales', y='NA_Sales');
```

<font color='green'>The scatter plot shows a positive relationship. The data is clustered togther, then gradually spreads diagonally to the top right, indicating that sales figures increase together in both regions.</font>

- <font color='blue'>Correlation coefficient</font>

```python
#Obtaining the numerical data only from the dataset
numerical_data = data.select_dtypes(include=['int64', 'float64'])

#generating the correlation coefficient for the chosen two features only
numerical_data[['NA_Sales','EU_Sales']].corr()
```

<font color='green'>The correlation coefficients is 0.767727, indicating a strong positive correlation between the two features. This suggests that as sales increase in North America, they tend to increase in Europe as well, and vice versa. This measure agrees with the observation from the scatter plot earlier.</font>

**Both tests agree with the initial assumption that the NA_Sales and EU_Sales of video games has a positive relation.**


### <font color='blue'>Are their features that can be used as index?</font>
